```python
from pyspark.sql import SparkSession
spark= SparkSession.builder.getOrCreate()
import pandas as pd

#File path
file_path= "/Volumes/workspace/default/airlines/On_Time_Reporting_Carrier_On_Time_Performance_(1987_present)_2023_1.csv"

#Read raw CSV - no changes, exactly as is 
raw_df = spark.read.csv(file_path, header=True, inferSchema=True)

print(f" Row loded : {raw_df.count(): ,}")
print(f"Columns: {len(raw_df.columns)}")
raw_df.printSchema()
```

```python
#Preview the data
display(raw_df.limit(10))
```

```python
#Save as Bronze Delta Tables
bronze_path = "/Volumes/workspace/default/airlines/bronze/flights"
raw_df.write.format("delta").mode("overwrite").save(bronze_path)
print("Bronze layer saved!!")

#Save as delta table
raw_df.write.format("delta").mode("overwrite").saveAsTable("bronze_flights")
print("Bronze layer saved as table!!")
```
